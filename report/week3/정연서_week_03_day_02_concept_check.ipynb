## 예제 1. math 모듈


```python
import math
radius = 3
area = math.pi * radius ** 2
print(round(area,2)) # round 모듈을 활용해서 28.27 로 출력해보기
```

## 예제 2. statistics 모듈


```python
import statistics
scores = [90, 72, 88, 95]
print('mean:',sum(scores)/len(scores)) # mean 출력1
print('mean:',statistics.mean(scores)) # mean 출력2
print('median:',statistics.median(scores)) # median 출력
```

## 예제 3. datetime 모듈


```python
from datetime import date, timedelta
today = date(2026, 6, 25)
next_week = today + timedelta(days=7)
print(today.isoformat()) # 오늘 날짜
print(next_week.isoformat()) # 다음주 날짜
```

## 예제 5. 작은 유틸리티 묶음


```python
def normalize_keyword(text):
    return text.strip().lower().replace(' ', '-')
keywords = [' Python ', 'API Practice', ' Report ']
print([normalize_keyword(word) for word in keywords])
```
