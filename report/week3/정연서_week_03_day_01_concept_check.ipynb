```python
print('START_WEEK03_DAY01_PROJECT_PLUS')
def greet(name):
    return f'{name}님, 환영합니다.'
print(greet('Mina')) # "Mina" 를 출력해주세요
```

## 예제 1. parameter와 return


```python
def add(a, b):
    return a + b
result = 3 + 4 # 3과 4를 더해서 결과를 출력해주세요.
print('result:', result)
```

## 예제 2. 기본값


```python
def format_label(name, level='beginner'):
    return f'{name}:{level}'
print(format_label('Mina'))
print(format_label('Jin','review'))
```

## 예제 3. 여러 값 반환


```python
def summarize(numbers):
    return min(numbers), max(numbers), sum(numbers)
low, high, total = summarize([3,7,10]) # [3, 7, 10] 을 입력해 보기
print(low, high, total)
```

## 예제 4. 함수 조합


```python
def add_tax(price, rate=0.1):
    return price * (1 + rate)
def make_price_card(name, price):
    return {'name': name, 'final_price': round(add_tax(price), 2)}
print(make_price_card('book', 12000)) # 'book', 12000 입력하기
```

## 예제 5. 작은 검증 함수


```python
def is_non_empty_text(value):
    return isinstance(value, str) and len(value.strip()) > 0
print(is_non_empty_text("T")) # True 출력하기
print(is_non_empty_text("")) # False 출력하기
```
