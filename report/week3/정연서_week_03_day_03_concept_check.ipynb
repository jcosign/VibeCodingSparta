# Week03 Day03 Concept Check Plus — HTTP, URL, JSON
API 호출 전에 URL, query parameter, status code, JSON 구조를 노트북 안에서 안전하게 연습합니다.


## 목표
- URL과 query parameter를 구분합니다.
- status code를 읽습니다.
- JSON 딕셔너리/리스트에서 값을 꺼냅니다.
- 필요한 key가 있는지 검증합니다.


```python
print('START_WEEK03_DAY03_PROJECT_PLUS')
from urllib.parse import urlencode
base_url =  "https://www.naver.com" # 자신이 원하는 URL 출력해보기 e.g https://example.com/books   /  https://www.yes24.com/Main/default.aspx
params = {'q': 'python', 'limit': 3}
url = base_url + '?' + urlencode(params)
print(url)
```

## 예제 1. status code 해석


```python
def explain_status(code):
    if code == 200:
        return 'success'
    if code == 404:
        return 'not found'
    return 'check needed'
for code in [200,404,500]: # 200, 404, 500 를 리스트로 범위로 만들어서 입력해보기
    print(code, explain_status(code))
```

## 예제 2. JSON 구조 읽기


```python
response = {'items': [{'title': 'Python Basics', 'authors': ['A']}, {'title': 'API Practice', 'authors': ['B']}]}
print(response['items'][0]['title']) # Python Basics 출력
print([item['title'] for item in response['items']]) # ['Python Basics', 'API Practice'] 출력
## 모든 title만 추출 후 리스트로 만든 것. / 해당 부분 보충 필요
```

## 예제 3. key 검증


```python
required = {'items'}
missing = required - set(response.keys())
print('missing:', missing)
print('has_items:', 'items' in response)
```

## 예제 4. 안전한 추출


```python
def extract_titles(payload):
    return [item.get('title', '제목 없음') for item in payload.get('items', [])]
print(extract_titles(response))
print(extract_titles({}))
```

## 예제 5. 로그 딕셔너리 만들기


```python
log = {'url': url, 'status_code': 200, 'row_count': len(response['items'])} # e.g log = {'url': url, 'status_code': 200, 'row_count': len(response['items'])}
print(log)
```
