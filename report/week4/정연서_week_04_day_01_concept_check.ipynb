# Week04 Day01 Project Check — DataFrame 구조를 직접 확인하기

이 노트북은 DataFrame을 만들고 `head`, `shape`, `columns`, `dtypes`를 차례대로 확인합니다. 마지막에는 본인이 값을 하나 바꿔 결과 문장을 다시 만듭니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# 출력 파일이 필요한 셀에서 사용할 폴더입니다.
# Colab에서도 이 폴더가 자동으로 만들어지므로 별도 파일을 받을 필요가 없습니다.
OUTPUT_DIR = Path('outputs')
OUTPUT_DIR.mkdir(exist_ok=True)
```

## 1. 작은 표 만들기
딕셔너리 목록을 DataFrame으로 바꾸면 행과 열을 가진 표가 됩니다.

```python
rows = [
    {'product': 'notebook', 'category': 'office', 'sales': 120, 'price': 3500},
    {'product': 'pen', 'category': 'office', 'sales': 80, 'price': 900},
    {'product': 'tumbler', 'category': 'life', 'sales': 150, 'price': 12000},
    {'product': 'bag', 'category': 'life', 'sales': 60, 'price': 24000},
]

# 행 목록을 표로 바꿉니다.
df = pd.DataFrame(rows)
print(df)
```

## 2. 앞부분과 뒷부분 확인
`head`와 `tail`은 전체를 대표하는 결론이 아니라 로딩 확인용입니다.

```python
print('앞의 2행')
print(pd.DataFrame(rows).head(2)) # head 2줄
print()
print('뒤의 2행')
print(df.tail(2)) # tail 2줄
```

## 3. 행과 열 개수 확인
`shape`는 `(행 수, 열 수)` 순서입니다.

```python
row_count, column_count = df.shape # df shape를 활용해서 row와 column 수를 확인하기
print('row_count:', row_count)
print('column_count:', column_count)
print(f'이 표는 {row_count}행 {column_count}열입니다.')
```

## 4. 열 이름과 자료형 확인
계산 전에 열 이름과 자료형을 보는 습관을 들입니다.

```python
print('columns:', list(df.columns)) # df 의 columns 를 List로 출력하기
print('dtypes:', df.dtypes.to_dict())
```

## 5. 필요한 열만 선택하기
보고 싶은 열만 골라 작은 표를 만들 수 있습니다.

```python
selected = df[['product', 'sales', 'price']] # 'product', 'sales', 'price' 열만 선택하기
print(selected)
```

## 6. 직접 바꿔보기
아래 `new_row`의 상품명이나 숫자를 바꾼 뒤 다시 실행합니다.

```python
new_row = {'product': 'planner', 'category': 'office', 'sales': 180, 'price': 4500}
expanded = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
print(expanded)
print('바뀐 shape:', expanded.shape)
```

## 7. 확인 문장 만들기
분석 결과는 문장으로 바뀌어야 합니다.

```python
message = f"상품 표에는 {expanded.shape[0]}개 행이 있고, sales 열의 최댓값은 {expanded['sales'].max()}입니다."
print(message)
print('PROJECT_DAILY_CHECK_DONE')
```
