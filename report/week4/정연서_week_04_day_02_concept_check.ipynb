# Week04 Day02 Project Check — 결측치, 필터링, 정렬

빈 값이 있는 표를 안전하게 확인하고, 조건에 맞는 행을 고른 뒤 정렬합니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# 출력 파일이 필요한 셀에서 사용할 폴더입니다.
# Colab에서도 이 폴더가 자동으로 만들어지므로 별도 파일을 받을 필요가 없습니다.
OUTPUT_DIR = Path('outputs')
OUTPUT_DIR.mkdir(exist_ok=True)
```

## 1. 결측치가 있는 표 만들기
`None`은 값이 없다는 뜻입니다. 0과 다르게 봐야 합니다.

```python
df = pd.DataFrame({
    'product': ['notebook', 'pen', 'tumbler', 'bag', 'planner'],
    'category': ['office', 'office', 'life', 'life', 'office'],
    'sales': [120, None, 150, 60, 95],
    'price': [3500, 900, 12000, None, 5000],
})
print(df)
```

## 2. 열별 결측 개수 확인
어떤 열이 비었는지 먼저 봅니다.

```python
missing_by_column = df.isna().sum()  # 결측 개수 확인하기
print(missing_by_column)
```

## 3. 결측 행만 따로 보기
결측치를 처리하기 전에 어떤 행인지 직접 확인합니다.

```python
missing_rows = df[df['category'].isna() | df['sales'].isna() | df['price'].isna()]     # 결측한 행만 따로 보기
print(missing_rows)
```

## 4. 사본에서 결측치 처리
원본을 바로 바꾸지 않고 `copy()`로 실험합니다.

```python
clean = df.copy()
clean['sales'] = clean['sales'].fillna(0)
clean['price'] = clean['price'].fillna(0) # 아래 출력과 동일한 결과내기
print(clean)
```

## 5. 조건으로 필터링하기
기준값 이상인 행만 남깁니다.

```python
threshold = 100
high_sales = clean[clean['sales'] >= 100 ] # 100이상의 값만 출력하기
print(high_sales[['product', 'sales']])
```

## 6. 정렬해서 우선순위 보기
값 자체를 바꾸는 것이 아니라 행 순서만 바뀝니다.

```python
sorted_df = clean.sort_values('sales', ascending=False) # 'sales', ascending=False 넣어보기
print(sorted_df[['product', 'sales']])
```

## 7. 직접 바꿔보기
`threshold`를 80 또는 130으로 바꿔 필터링 결과가 어떻게 달라지는지 봅니다.

```python
my_threshold = 80
selected = clean[clean['sales'] >= my_threshold]
print('기준:', my_threshold)
print(selected[['product', 'sales']])
print(f'선택된 행 수는 {len(selected)}개입니다.')
print('PROJECT_DAILY_CHECK_DONE')
```
