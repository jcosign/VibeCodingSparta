# Week04 Day04 Project Check — 시각화와 해석

요약표를 그래프로 바꾸고 확인한 사실과 다음 확인을 문장으로 남깁니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# 출력 파일이 필요한 셀에서 사용할 폴더입니다.
# Colab에서도 이 폴더가 자동으로 만들어지므로 별도 파일을 받을 필요가 없습니다.
OUTPUT_DIR = Path('outputs')
OUTPUT_DIR.mkdir(exist_ok=True)
```

## 1. 그래프용 표 만들기

```python
df = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'sales': [120, 150, 90, 170, 210],
    'visits': [300, 330, 260, 390, 420],
})
print(df)
```

## 2. 막대그래프로 크기 비교
범주별 크기를 비교할 때 막대그래프가 읽기 쉽습니다.

```python
ax = df.plot(kind='bar', x='day', y='sales', legend=False, title='Sales by day')
# ax = df.plot 을 활용해서 막대그래프 그리기
ax.set_xlabel('day')
ax.set_ylabel('sales')
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'week04_day04_sales_bar.png')
plt.close()
print('saved:', OUTPUT_DIR / 'week04_day04_sales_bar.png')
```

## 3. 선그래프로 흐름 보기
시간이나 순서가 있는 값은 선그래프가 자연스럽습니다.

```python
ax = df.plot(kind='line', x='day', y=['sales', 'visits'], marker='o', title='Sales and visits')
# ax = df.plot을 활용해서 선그래프 그리기
ax.set_xlabel('day')
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'week04_day04_line.png')
plt.close()
print('saved:', OUTPUT_DIR / 'week04_day04_line.png')
```

## 4. 가장 큰 값 찾기

```python
top_day = df.sort_values('sales', ascending=False).iloc[0] # 가장 큰값 출력해보기
print(top_day)
print()
print('가장 큰 sales:', top_day.to_dict())
```

## 5. 해석 문장 작성
확인한 사실, 가능한 해석, 다음 확인을 분리합니다.

```python
fact = f"가장 큰 sales는 {top_day['day']}의 {top_day['sales']}입니다."
possible = '주 후반 수요가 높아졌을 가능성이 있습니다.'
next_check = '이벤트, 할인, 요일별 표본 수를 추가로 확인해야 합니다.'
print(fact)
print(possible)
print(next_check)
```

## 6. 직접 바꿔보기
새로운 하루 값을 추가한 뒤 그래프 저장 결과를 다시 만듭니다.

```python
extra = pd.DataFrame([{'day': 'Sat', 'sales': 190, 'visits': 410}])
expanded = pd.concat([df, extra], ignore_index=True)
ax = expanded.plot(kind='bar', x='day', y='sales', legend=False, title='Sales by day with Saturday')
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'week04_day04_sales_bar_with_sat.png')
plt.close()
print(expanded)
print()
print('PROJECT_DAILY_CHECK_DONE')
```
