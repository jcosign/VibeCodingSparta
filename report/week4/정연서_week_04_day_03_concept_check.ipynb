# Week04 Day03 Project Check — groupby와 집계

여러 행을 범주별로 묶고 합계, 평균, 개수를 계산합니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# 출력 파일이 필요한 셀에서 사용할 폴더입니다.
# Colab에서도 이 폴더가 자동으로 만들어지므로 별도 파일을 받을 필요가 없습니다.
OUTPUT_DIR = Path('outputs')
OUTPUT_DIR.mkdir(exist_ok=True)
```

## 1. 집계용 표 만들기
범주와 숫자 열이 함께 있어야 groupby 연습이 쉽습니다.

```python
df = pd.DataFrame({
    'category': ['office', 'office', 'life', 'life', 'life', 'digital'],
    'product': ['notebook', 'pen', 'tumbler', 'bag', 'cup', 'keyboard'],
    'sales': [120, 80, 150, 60, 90, 110],
    'price': [3500, 900, 12000, 24000, 7000, 45000],
})
print(df)
```

## 2. 범주별 합계
가장 기본적인 groupby는 범주별 합계입니다.

```python
sales_by_category = df.groupby('category')['sales'].sum().reset_index() # groupby를 이용해서 범주를 출력해보기 이때, sum().reset_index() 활용하기
# category는 묶는 열이고, sales는 계산하는 열
print(sales_by_category)
```

## 3. 여러 집계를 한 번에 계산
합계, 평균, 개수를 함께 보면 해석이 안정적입니다.

```python
summary = df.groupby('category').agg(
    sales_total=('sales', 'sum'),
    sales_mean=('sales', 'mean'),
    product_count=('product', 'count'),
).reset_index()
print(summary)
```

## 4. 범주별 비중 계산
합계만 보지 말고 전체에서 차지하는 비율도 확인합니다.

```python
total_sales = summary['sales_total'].sum()
summary['sales_share'] = (summary['sales_total'] / total_sales).round(3) # 범주별로 비율 이용해보기 e.g   (summary['sales_total'] / total_sales).round(3)
print(summary)
```

## 5. value_counts로 개수 빠르게 보기
한 열의 범주별 개수는 `value_counts()`가 편합니다.

```python
print(df['category'].value_counts())
```

## 6. 직접 바꿔보기
새 상품 행을 하나 추가하고 요약표가 어떻게 바뀌는지 확인합니다.

```python
new_item = pd.DataFrame([{'category': 'digital', 'product': 'mouse', 'sales': 70, 'price': 18000}])
expanded = pd.concat([df, new_item], ignore_index=True)
new_summary = expanded.groupby('category')['sales'].sum().reset_index()
print(new_summary)
```

## 7. 해석 문장 만들기
가장 큰 범주를 찾고 한 문장으로 씁니다.

```python
top = new_summary.sort_values('sales', ascending=False).iloc[0]
print(f"가장 큰 범주는 {top['category']}이며 sales 합계는 {top['sales']}입니다.")
print('PROJECT_DAILY_CHECK_DONE')
```
