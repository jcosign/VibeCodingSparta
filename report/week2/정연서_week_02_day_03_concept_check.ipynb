# Week02 Day03 Concept Check Plus — 조건문과 반복문
조건문과 반복문은 데이터를 고르고, 누적하고, 중단하고, 다시 확인하는 기본 흐름입니다.


## 목표
- `if/elif/else`로 상태를 나눕니다.
- `for`로 값을 하나씩 처리합니다.
- `while`로 조건이 유지되는 동안 반복합니다.
- `break`, `continue`, comprehension을 읽습니다.


```python
print('START_WEEK02_DAY03_PROJECT_PLUS')
score = 84

# 90점 이상은 A , 80점 이상은 B 로 출력되게 만들어주세요.
if score >= 90 :
    grade = 'A'
elif score :
    grade = 'B'
else:
    grade = 'review'
print('등급:', grade)
```

## 예제 1. 조건 순서
조건은 위에서 아래로 검사합니다. 넓은 조건을 먼저 쓰면 세부 조건이 가려질 수 있습니다.


```python
value = 92
if value >= 60: # 60점 이상은 pass 로 출력되게 만들어주세요.
    print('잘못된 순서: pass')

if value >= 90: # 90점 이상은 pass, 60-90점 이상은 passs 로 출력되게 만들어주세요.
    print('올바른 순서: pass')
elif value >= 60:
    print('passs')
```

## 예제 2. for 반복과 누적
여러 값을 처리하고 합계나 개수를 만들 때 씁니다.


```python
scores = [92, 75, 48, 88, 61]
passed = []
review = []
for score in scores : # scores를 for문에 넣는 방법은?
    if score >= 70:
        passed.append(score)
    else:
        review.append(score)
print('통과:', passed)
print('재확인:', review)
```

## 예제 3. while 반복과 안전장치
무한 반복을 막기 위해 조건이 변하는 줄을 반드시 둡니다.


```python
count = 0
while count < 4 : # count가 4미만일 때까지만 while 문이 실행되게 해주세요.
    print('현재 count:', count)
    count += 1 # count는 1씩 증가되게 해주세요
```

## 예제 4. break와 continue
처리할 값과 건너뛸 값을 구분합니다.


```python
numbers = [3, 5, -1, 7, 0, 9]
used = []
for number in numbers:
    if number < 0 : # 0 미만은 continue
        continue
    if number == 0 : # 0 이면 for문을 중지해주세요.
        break
    used.append(number)
print('사용한 값:', used)
```

## 예제 5. comprehension
짧은 변환은 한 줄로 쓸 수 있지만, 처음에는 for문으로 먼저 이해합니다.


```python
raw_scores = [45, 75, 95, 60, 87]
corrected = [min(score + 5, 100) for score in raw_scores]
print('보정 전:', raw_scores)
print('보정 후:', corrected)
```
