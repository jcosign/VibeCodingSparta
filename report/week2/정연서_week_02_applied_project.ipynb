# Week02 Applied Project Plus — 학습 체크리스트 처리기
리스트, 딕셔너리, 셋, 조건문, 반복문, 예외처리를 한 번에 묶어 작은 제출 점검기를 만듭니다.


## 목표
- 여러 학생 제출 기록을 리스트와 딕셔너리로 표현합니다.
- 제출 개수와 TODO 수정 여부를 조건문으로 판정합니다.
- 잘못된 점수 입력은 예외처리로 분리합니다.
- 최종 요약 딕셔너리를 출력합니다.


```python
print('START_WEEK02_APPLIED_PROJECT_PLUS')
submissions = [
    {'name': 'Mina', 'notebooks': 3, 'edited': True, 'score': '90'},
    {'name': 'Lisa', 'notebooks': 1, 'edited': False, 'score': '80'},
    {'name': 'Anna', 'notebooks': 4, 'edited': True, 'score': '없음'},
    {'name': 'John', 'notebooks': 2, 'edited': True, 'score': '95'}
# 4개의 name , notebook, edited , score를 가진 dictionary를 component를 추가해주세요.
]
print('row_count:', len(submissions))
```

## 1. 제출 상태 판정


```python
for row in submissions:
    if row['notebooks'] >= 3 and row['edited'] == True: # notebook이 3 이상 그리고 edited 일 때
        row['status'] = 'ready' # status 는 ready
    elif row['notebooks'] >= 2: # notebook이 2 이상
        row['status'] = 'review' # status 는 review
    else:
        row['status'] = 'needs_help'
print(submissions)
```

## 2. 점수 변환과 실패 기록


```python
valid_scores = []
failed_scores = []
for row in submissions:
    try:
        valid_scores.append(int(row['score'])) # valid_scores에 'score'를 int dtype으로 append 해주세요.
    except ValueError:
        failed_scores.append({'name': row['name'], 'raw_score': row['score']})
print('valid_scores:', valid_scores)
print('failed_scores:', failed_scores)
```

## 3. 셋으로 상태 종류 확인


```python
status_set = {row['status'] for row in submissions}
print('status_set:', status_set)
print('ready_count:', sum(1 for row in submissions if row['status'] == 'ready'))
```

## 4. 요약 만들기


```python
summary = {
    'student_count': len(submissions),
    'ready': sum(1 for row in submissions if row['status'] == 'ready'),
    'review': sum(1 for row in submissions if row['status'] == 'review'),
    'needs_help': sum(1 for row in submissions if row['status'] == 'needs_help'),
    'score_average': round(sum(valid_scores) / len(valid_scores), 2) if valid_scores else 0,
    'failed_score_count': len(failed_scores),
}
print(summary)
```

## 5. 직접 바꿔보기
아래 기준을 바꾸면 ready 판정이 달라집니다.


```python
required_notebooks = 3  # TODO: 2 또는 4로 바꿔보세요.
ready_names = [row['name'] for row in submissions if row['notebooks'] >= required_notebooks and row['edited']]
print('required_notebooks:', required_notebooks)
print('ready_names:', ready_names)
```
