# Week02 Day02 Concept Check Plus — 딕셔너리와 셋
딕셔너리는 이름표가 붙은 값, 셋은 중복 없는 묶음입니다. key 접근, `.get()`, 반복, 셋 연산까지 확인합니다.


## 목표
- 딕셔너리 key로 값을 꺼냅니다.
- 없는 key를 안전하게 다룹니다.
- `.items()` 반복으로 key/value를 함께 봅니다.
- 셋으로 중복 제거와 교집합/차집합을 확인합니다.


```python
print('START_WEEK02_DAY02_PROJECT_PLUS')
profile = {'name': 'Mina', 'level': 'beginner', 'tool': 'Colab'}
print(profile["name"]) # Mina 를 출력하기
print(profile.get('city', '도시없음')) # "도시 없음" 이라는 메세지를 에러 메세지로 출력하기
```

## 예제 1. 딕셔너리 수정
값을 추가하거나 바꾸면 같은 key의 값이 갱신됩니다.


```python
profile['level'] = 'review' # TODO: Dictionary에 'level' 이라는 key를 가지고 있으며, value는 'review' 입니다.
profile['week'] = 2 # TODO: Dictionary에 'week' 이라는 key를 가지고 있으며, value는 '2' 입니다.
print(profile)
```

## 예제 2. key와 value 반복
설정값이나 응답 데이터를 읽을 때 자주 쓰는 패턴입니다.


```python
settings = {'font': 'large', 'theme': 'light', 'autosave': True}
for key, value in settings.items(): #HINT: item / items 사용하기
    print(key, '=>', value)
```

## 예제 3. 셋으로 중복 제거
중복된 응답, 태그, 관심사를 정리할 때 유용합니다.


```python
tags = ['python', 'list', 'python', 'dict', 'set', 'dict']
unique_tags = set(tags) # set을 활용하세요
print('중복 제거:', unique_tags)
print('정렬 보기:', sorted(unique_tags))
```

## 예제 4. 셋 연산
두 날의 학습 키워드를 비교합니다.


```python
day1 = {'list', 'tuple', 'index'}
day2 = {'dict', 'set', 'index'}
print('공통:', day1 & day2)
print('Day2에 새로 나온 것:', day2 - day1)
print('전체:', day1 | day2)
```

## 예제 5. 빈도 세기
딕셔너리로 값이 몇 번 나오는지 셀 수 있습니다.


```python
words = ['list', 'dict', 'list', 'set', 'dict', 'dict']
counts = {}
for word in words: # HINT: range에 list를 넣어보세요
    counts[word] = counts.get(word, 0) + 1
print(counts)
```
