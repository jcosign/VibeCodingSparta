# Week02 Day04 Concept Check Plus — 예외처리
예외처리는 오류를 숨기는 기술이 아니라, 실패할 수 있는 입력을 안전하게 다루는 기술입니다.


## 목표
- `try/except`로 변환 실패를 처리합니다.
- 실패 값과 성공 값을 나누어 기록합니다.
- `.get()`으로 key 누락을 방어합니다.
- `else/finally` 흐름을 읽습니다.


```python
print('START_WEEK02_DAY04_PROJECT_PLUS')
value = '20'
try:
    number = int(value)
    print('변환 성공:', number)
except ValueError : # Value가 잘못되었을 때 출력해주세요.
    print('변환 실패')
```

## 예제 1. 여러 값 변환하기
실제 데이터에는 숫자로 바뀌지 않는 값이 섞일 수 있습니다.


```python
values = ['10', 'bad', '30', 'x']
success = []
failed = []
for value in values:
    try:
        success.append(int(value))
    except ValueError : # Value가 잘못되었을 때 출력해주세요.
        failed.append(value)
print('성공:', success)
print('실패:', failed)
```

## 예제 2. else와 finally
`else`는 오류가 없을 때, `finally`는 성공/실패와 상관없이 실행됩니다.


```python
raw = '42'
try:
    parsed = int(raw)
except ValueError : # Value가 잘못되었을 때 출력해주세요.
    print('실패')
else : # 성공시에만 출력되게 해주세요.
    print('성공:', parsed)
finally : # 항상 출력되게 해주세요.
    print('변환 시도 종료')
```

## 예제 3. 딕셔너리 key 방어
없는 key를 바로 꺼내면 오류가 나므로 `.get()`을 사용할 수 있습니다.


```python
user = {'name': 'Mina'}
print('도시:', user.get('city', '도시없음')) # 'city' 라는 dictionary key를 검색하고, 없으면 '도시 없음'을 출력하세요
```

## 예제 4. 안전한 평균 계산
빈 목록, 문자열 값, 0으로 나누기 같은 상황을 방어합니다.


```python
raw_scores = ['90', '없음', '72', '55']
valid = []
failed = []
for raw in raw_scores:
    try:
        valid.append(int(raw))
    except : # Value가 잘못되었을 때 출력해주세요.
        failed.append(raw)
average = sum(valid) / len(valid) if valid else 0
print('valid:', valid)
print('failed:', failed)
print('average:', round(average, 2))
```

## 예제 5. 작은 검증 함수
예외처리는 함수로 묶으면 반복 사용하기 좋습니다.


```python
def parse_score(raw):
    try:
        return {'ok': True, 'value': int(raw)}
    except ValueError: # Value가 잘못되었을 때 출력해주세요.
        return {'ok': False, 'value': None, 'raw': raw}

print(parse_score('100'))
print(parse_score('결석'))
```
