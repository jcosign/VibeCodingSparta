## 예제 1. 인덱싱과 슬라이싱
위치 번호는 0부터 시작합니다. 슬라이싱은 시작은 포함하고 끝은 포함하지 않습니다.


```python
scores = [60, 75, 88, 92, 70, 81]
print('앞 세 개:', scores[0:3])
print('가운데:', scores[2:5])
print('두 칸씩:', scores[::2])
print('역순:', scores[::-1])
```

## 예제 2. 리스트 값 추가와 제거
프로젝트에서는 학생이 직접 값을 하나 더 넣고 출력 변화를 확인해야 합니다.


```python
tasks = ['개념 읽기', '예제 실행']
tasks.append('출력설명') # '출력 설명' 추가하기
tasks.insert(1, '주석확인') # index 1에 '주석 확인' 추가
# +++ 숫자 많이 쓰면 맨 뒤에 추가 됨
removed = tasks.pop(-1) # pop으로 마지막 출력만 꺼내기
print('현재 tasks:', tasks)
print('꺼낸 값:', removed)
```

## 예제 3. 정렬과 원본 보존
`sorted()`는 새 목록을 만들고, `sort()`는 원본을 바꿉니다.


```python
raw = [82, 95, 70, 88]
ascending = sorted(raw)
raw.sort(reverse=True) # Question: True or False ?
print('새 정렬:', ascending)
print('원본 변경:', raw)
```

## 예제 4. 튜플 언패킹
튜플은 좌표, 범위, 고정된 묶음 값을 표현할 때 좋습니다.


```python
point = (3, 7)
x, y = point   # point를 x와 y로 2개로 나눠서 출력하기
print('x:', x)
print('y:', y)
print('좌표 합:', x + y)
```

## 예제 5. 중첩 리스트
표처럼 생긴 데이터를 읽는 첫 연습입니다. 나중에 DataFrame을 볼 때 도움이 됩니다.


```python
rows = [['Mina', 90], ['Jin', 75], ['Ara', 88]]
print('두 번째 행:', rows[1])
print('세 번째 이름:', rows[2][0])
print('점수 목록:', [row[1] for row in rows])
```
