# 04. 미니 프로필 카드 만들기

마지막 노트북은 변수, 리스트, 딕셔너리, 출력문을 한 번에 묶는 쉬운 종합 연습입니다.

## 1. 기본 값 정하기

```python
# TODO 1: 아래 값만 본인에게 맞게 바꿔보세요.
student_name = "정연서" # 본인 성함
learning_goal = "Ai를 활용하여 반복&단순 업무 줄이기" # 본인 목표
favorite_examples = ["LLM", "Python 실무", "반복데이터 단순화"] # 가장 다뤄졌으면 하는 주제 3가지 적기e.g "LLM"
```

## 2. 딕셔너리로 묶기

```python
# TODO 2: profile 딕셔너리에 "memo" 값을 하나 더 추가해보세요.
profile = {
    "name": student_name,
    "goal": learning_goal,
    "examples": favorite_examples,
    "memo": "셀을 위에서 아래로 실행한다",
}
```

## 3. 카드 문장 만들기

```python
# TODO 3: card_lines에 출력하고 싶은 문장 하나를 더 추가해보세요.
card_lines = [
    f"이름: {profile['name']}",
    f"목표: {profile['goal']}",
    f"예제: {', '.join(profile['examples'])}",
    f"메모: {profile['memo']}",
]
```

## 4. 카드 출력하기

```python
print("=" * 30)
print("Week01 Easy Project Card")
print("=" * 30)
for line in card_lines:
    print(line)
print("=" * 30)
```

## 5. 중복 예제 확인

좋아하는 예제 목록에 중복이 있어도 집합으로 정리할 수 있습니다.

```python
duplicated_examples = favorite_examples + ["print", "리스트"]
unique_examples = set(duplicated_examples)
print("중복 포함:", duplicated_examples)
print("중복 제거:", unique_examples)
```
