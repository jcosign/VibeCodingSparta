# 03. 딕셔너리와 집합 복습

이번 노트북은 key-value 구조와 중복 제거를 연습합니다.

## 1. 딕셔너리 만들기

딕셔너리는 순서보다 이름이 중요한 값 묶음입니다.

```python
# TODO 1: profile의 name, role, goal 값을 본인에게 맞게 바꿔보세요.
profile = {
    "name": "정연서", # 수강생이름
    "role": "데이터분석사", # 직무
    "goal": "Ai를 활용하여 반복단순업무를 줄일 것", # 목표
}
print(profile)
```

## 2. key로 값 꺼내기

딕셔너리는 숫자 위치가 아니라 key 이름으로 값을 가져옵니다.

```python
print("이름:", profile["name"]) # profile내의 "name" 부르기
print("목표:", profile["goal"]) # profile내의 "goal" 부르기
```

## 3. 딕셔너리에 값 추가하기

```python
# TODO 2: 새로운 key-value를 하나 추가해보세요.
profile["favorite_example"] = "print 연습" # "favorite_example" 이라는 새로운 Key에 "print 연습" 이라는 value 넣기
print(profile)
```

## 4. 집합으로 중복 제거하기

```python
# TODO 3: interests에 중복 값을 하나 넣고, set으로 중복 제거를 확인하세요.
interests = ["자동화", "데이터", "자동화", "리포트"]
unique_interests =  set(interests)  # interests 변수를 set을 활용해서 새로운 unique_interests 만들기
print("원래 리스트:", interests)
print("중복 제거:", unique_interests)
```

```python
completion_log = {
    "notebook": "03_dictionary_set",
    "status": "PROJECT_EASY_DONE",
    "profile_keys": list(profile.keys()),
    "unique_interest_count": len(unique_interests),
}
print(completion_log)
```
