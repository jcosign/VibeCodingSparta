# 02. 리스트와 튜플 복습

이번 노트북에서는 여러 값을 순서대로 묶는 방법을 연습합니다.

## 1. 리스트 만들기

리스트는 오늘 한 일, 관심 주제, 파일 이름처럼 여러 값을 순서대로 담을 때 씁니다.

```python
# TODO 1: tasks 리스트에 오늘 해본 일을 1개 더 추가해보세요.
tasks = ["리스트 공부하기", "변수 생성하기", "보충공부하기"]
tasks.append("변수 바꿔보기")
print(tasks)
```

## 2. 인덱스로 값 꺼내기

컴퓨터는 첫 번째 위치를 0번으로 봅니다.

```python
# TODO 3: 두 번째 항목도 출력해보세요. 힌트: 두 번째는 [1]입니다.
print("첫 번째 할 일:", tasks[0])
print("두 번째 할 일:", tasks[1])
```

## 3. 튜플 만들기

튜플은 순서가 정해져 있고 바꾸지 않을 값에 사용합니다.

```python
# TODO 2: tuple 안의 값은 직접 바꾸지 말고, 새 tuple을 만들어보세요.
fixed_steps = ("열기", "실행", "확인")
my_steps = fixed_steps + ("저장",)
print("기본 순서:", fixed_steps) # "열기", "실행", "확인"
print("내가 정리한 순서:", my_steps) # "열기", "실행", "확인", "저장"
```

## 4. 리스트 길이 확인

`len()`은 값이 몇 개인지 알려줍니다.

```python
print("tasks 개수:", len(tasks)) # tasks 변수 길이
print("fixed_steps 개수:", len(fixed_steps)) # fixed_steps 변수 길이
```
